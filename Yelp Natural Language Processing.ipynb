```python
import json

restaurant_ids = []

# open the business file
with codecs.open(business_filepath, encoding='utf_8') as f:
    
    # iterate through each line (json record) in the file
    for business_json in f:
        
        #convert the json record to a python dict
        business= json.loads(business_json)
        
        # if this business is not a resturant, skip to the next one
        if u'Restaurants' not in business[u'categories']:
            continue
            
        # add the restaurant business id to out restaurant_ids set
        restaurant_ids.append(business[u'business_id'])
        
# turn restaurant_ids into a frozenset, as we don't need to change it.
restaurant_ids=restaurant_ids

# print the number of unique restaurant ids in the dataset
print('{:,}'.format(len(restaurant_ids)), u'restaurants in the dataset.')
```
